### Time Summary

```python
# List of file paths for the uploaded CSV files
file_paths = [
    '/content/drive/MyDrive/ADB-Dataware-housing-with-Google-bigquery-and-snowflake-main/big_query/BigQueryLoadTime/TPCH_05_loading_times.csv',
    '/content/drive/MyDrive/ADB-Dataware-housing-with-Google-bigquery-and-snowflake-main/big_query/BigQueryLoadTime/TPCH_1G_loading_times.csv',
    '/content/drive/MyDrive/ADB-Dataware-housing-with-Google-bigquery-and-snowflake-main/big_query/BigQueryLoadTime/TPCH_2G_loading_times.csv',
    '/content/drive/MyDrive/ADB-Dataware-housing-with-Google-bigquery-and-snowflake-main/big_query/BigQueryLoadTime/TPCH_3G_loading_times.csv'
]

# Read each CSV file and store them in a list
datasets = [pd.read_csv(file) for file in file_paths]

# Display the structure of the first dataset to understand how to combine them
datasets[0].head()

# Create a new DataFrame to hold all the combined data with an additional 'File Size' column
combined_data = pd.DataFrame()

# Loop over each dataset and append to the combined_data DataFrame
for i, dataset in enumerate(datasets):
    dataset['File Size'] = f"{i}G" if i > 0 else "0.5G"  # Assign file size based on the order
    combined_data = combined_data.append(dataset, ignore_index=True)

# Set the order of file sizes to maintain consistency in the plot
file_size_order = ["0.5G", "1G", "2G", "3G"]

pivot_data = combined_data.pivot(index='Dataset', columns='File Size', values='Loading Time')

# Define a simple color scheme
color_scheme = ['#A6CEE3', '#1F78B4', '#B2DF8A', '#33A02C']

# Plot
fig, ax = plt.subplots(figsize=(14, 10))

# We stack the bars using a cumulative sum to place them on top of each other
bottom = np.zeros(len(pivot_data))

# Loop to plot the bars
for i, col in enumerate(pivot_data.columns[::-1]):  # Reverse the order for proper stacking
    ax.bar(pivot_data.index, pivot_data[col], bottom=bottom, label=col, color=color_scheme[i])
    bottom += pivot_data[col].values

# Adding the title and labels with a more academic look
ax.set_title('Loading Times by Dataset and File Size', fontsize=20, weight='bold')
ax.set_xlabel('Dataset', fontsize=16, weight='bold')
ax.set_ylabel('Loading Time (seconds)', fontsize=16, weight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', rotation_mode='anchor', weight='bold')

# Adding a legend with a clear label
ax.legend(title='File Size', title_fontsize='13', loc='upper right', fontsize='12')

# Set the background to white with no gridlines for a clean look
ax.set_facecolor('white')
ax.grid(False)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Load the remaining three files as separate datasets
execution_time_file_1 = '/content/drive/MyDrive/ADB-Dataware-housing-with-Google-bigquery-and-snowflake-main/big_query/Throughput_Bigquery/Throughput_time_cu4/PUBLIC/execution_times.csv'
execution_time_file_2 = '/content/drive/MyDrive/ADB-Dataware-housing-with-Google-bigquery-and-snowflake-main/big_query/Throughput_Bigquery/Throughput_time_cu4/TPCH_1G/execution_times.csv'  # Second file
execution_time_file_3 = '/content/drive/MyDrive/ADB-Dataware-housing-with-Google-bigquery-and-snowflake-main/big_query/Throughput_Bigquery/Throughput_time_cu4/TPCH_2G/execution_times.csv'  # Third file
execution_time_file_4 = '/content/drive/MyDrive/ADB-Dataware-housing-with-Google-bigquery-and-snowflake-main/big_query/Throughput_Bigquery/Throughput_time_cu4/TPCH_3G/execution_times.csv'  # Fourth file

data1 = pd.read_csv(execution_time_file_1)
data1['Dataset'] = 'TPCH_05G'

# Load the second file and add an identifier
data2 = pd.read_csv(execution_time_file_2)
data2['Dataset'] = 'TPCH_1G'

# Load the third file and add an identifier
data3 = pd.read_csv(execution_time_file_3)
data3['Dataset'] = 'TPCH_2G'

# Load the fourth file and add an identifier
data4 = pd.read_csv(execution_time_file_4)
data4['Dataset'] = 'TPCH_3G'

# Combine all datasets into one DataFrame
combined_execution_data = pd.concat([data1, data2, data3, data4], ignore_index=True)

# Check for any negative or extremely large run times in the combined data
negative_times = combined_execution_data[combined_execution_data['Run_Time'] < 0]
extreme_large_times = combined_execution_data[combined_execution_data['Run_Time'] > combined_execution_data['Run_Time'].quantile(0.99)]

# Display the count of negative times and extreme large times, and a few examples if they exist
negative_count = negative_times.shape[0]
extreme_large_count = extreme_large_times.shape[0]

(negative_count, negative_times.head(), extreme_large_count, extreme_large_times.head())


# Removing extreme large values above the 99th percentile for visualization
cutoff = combined_execution_data['Run_Time'].quantile(0.99)
filtered_data = combined_execution_data[combined_execution_data['Run_Time'] <= cutoff]

# Now let's create a boxplot for the filtered data
plt.figure(figsize=(14, 8))
boxplot = sns.boxplot(x='SQL_File', y='Run_Time', hue='Dataset', data=filtered_data, palette="muted")

# Set the y-axis to start at 0 to avoid negative values
boxplot.set_ylim(bottom=0)

# Adding the title and labels
plt.title('Throughput Time Comparison Across Datasets (Filtered)', fontsize=20)
plt.xlabel('SQL File', fontsize=16)
plt.ylabel('Throughput Time (seconds)', fontsize=16)

# Show the plot
plt.tight_layout()
plt.show()
```
